```python
testFile.dtypes
```

```python
testFile.head()
```

```python
newFile = testFile.dropna()
```

```python
new_file
```

```python
newfile2  = new_file.set_index("PassengerId")
```

```python
#This is to show how fare is distributed for different passenger id
newfile2[['Fare']].plot(kind='bar')
```

### Pie Chart

Plotting a pie chart for each passenger name for their fares. (Considering only first 10 records)

```python
new_file2 = new_file.set_index("Name")
```

```python
new_file2.plot.pie(y='Fare',figsize= (5,5))
```
